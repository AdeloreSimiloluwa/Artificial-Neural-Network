## Build the Components

```python
# Import Kubeflow SDK
import kfp
import kfp.dsl as dsl
import kfp.components as comp


import pickle
import tensorflow as tf
from tensorflow.python import keras
```

```python
# Download the dataset and split into training and test data. 
DATA_DIR = 'data/'
data = pd.read_csv(DATA_DIR + "Churn_Modelling.csv")
X_train,X_test,y_train,y_test = train_test_split( X,y, test_size=0.2, random_state = 0)
# Normalize the data.
sc =StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

y_train = y_train.values
y_test = y_test.values
def train(data_path, model_file):


    # Define the model using Keras.
    classifier = keras.models.Sequential()
    classifier.add(keras.layers.Dense(units =6, activation='relu', input_dim=11))
    classifier.add(keras.layers.Dense(units =6, activation='relu'))
    classifier.add(keras.layers.Dense(units =1, activation='sigmoid'))

    classifier.summary()

    classifier.compile(optimizer = 'adam', loss='binary_crossentropy' 
                    , metrics =['accuracy'])

    return classifier
```

```python
classifier = train("", "")
# # Run a training job with specified number of epochs
classifier.fit(X_train, y_train, batch_size=10 , epochs=2)
```

```python
#Evaluate the model and print the results
test_loss, test_acc = classifier.evaluate(X_test,  y_test, verbose=0)
print('Test accuracy:', test_acc)

#Save the model to the designated 
classifier.save(os.path.join("trained_model"))

#Save the test_data as a pickle file to be used by the predict component.
with open(os.path.join("test_data"), 'wb') as f:
    pickle.dump((X_test,  y_test), f)
```

```python
def predict(data_path, model_file, churn_rate):
    
    # func_to_container_op requires packages to be imported inside of the function.
    import pickle

    import tensorflow as tf
    from tensorflow import keras

    import numpy as np
    
    # Load the saved Keras model
    classifier = keras.models.load_model("trained_model")

    # Load and unpack the test_data
    with open(os.path.join("test_data"),'rb') as f:
        test_data = pickle.load(f)
    # Separate the test_images from the test_labels.
    X_test,  y_test = test_data


    # See https://github.com/kubeflow/pipelines/issues/2320 for explanation on this line.
    churn_rate = int(churn_rate)

    # make predictions.
    y_pred = classifier.predict(X_test)


    # create a threshold for the confution matrics
    y_pred=(y_pred>0.5)
    
    with open(f'result.txt', 'w') as result:
        result.write(" Prediction: {} ".format(y_pred))
    
    print('Prediction has be saved successfully!')
```

```python
# Create train and predict lightweight components.
train_op = comp.func_to_container_op(train, base_image='tensorflow/tensorflow:latest-gpu-py3')
predict_op = comp.func_to_container_op(predict, base_image='tensorflow/tensorflow:latest-gpu-py3')
```

## Build Kubeflow Pipeline

```python
#Create a client to enable communication with the Pipelines API server.
client = kfp.Client()
```

```python
# Define the pipeline
@dsl.pipeline(
   name='Customer Churn Pipeline',
   description='An ML pipeline that performs Customer Churn model training and prediction.'
)

# Define parameters to be fed into pipeline
def churn_container_pipeline(
    data_path: str,
    model_file: str, 
    churn_rate: int
):
    
    # Define volume to share data between components.
    vop = dsl.VolumeOp(
    name="create_volume",
    resource_name="data-volume", 
    size="1Gi", 
    modes=dsl.VOLUME_MODE_RWM)
    
    # Create churn training component.
    churn_training_container = train_op(data_path, model_file) \
                                    .add_pvolumes({data_path: vop.volume})

    # Create Churn prediction component.
    churn_predict_container = predict_op(data_path, model_file, churn_rate) \
                                    .add_pvolumes({data_path: churn_training_container.pvolume})
    
    # Print the result of the prediction
    mnist_result_container = dsl.ContainerOp(
        name="print_prediction",
        image='library/bash:4.4.23',
        pvolumes={data_path: churn_predict_container.pvolume},
        arguments=['cat', f'{data_path}/result.txt']
    )
```

## Run the Pipeline

```python
DATA_PATH = '/churn'
MODEL_PATH='churn_classifier.h5'
# An integer representing an image from the test set that the model will attempt to predict the label for.
CHURN_RATE = 3
```

```python
pipeline_func = churn_container_pipeline
```

```python
experiment_name = 'costumer_churn_kubeflow'
run_name = pipeline_func.__name__ + ' run'

arguments = {"data_path":DATA_PATH,
             "model_file":MODEL_PATH,
             "churn_rate": CHURN_RATE}

# Compile pipeline to generate compressed YAML definition of the pipeline.
kfp.compiler.Compiler().compile(pipeline_func,  
  '{}.zip'.format(experiment_name))

# Submit pipeline directly from pipeline function
run_result = client.create_run_from_pipeline_func(pipeline_func, 
                                                  experiment_name=experiment_name, 
                                                  run_name=run_name, 
                                                  arguments=arguments)
```
